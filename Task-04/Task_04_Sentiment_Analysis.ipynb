# Task 4: Sentiment Analysis on Social Media Data

Analyze and visualize sentiment patterns in social media data to understand public opinion and attitudes towards specific topics or brands.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import warnings

warnings.filterwarnings('ignore')
plt.style.use('ggplot')
```

## 1. Load the Dataset

```python
df = pd.read_csv('data-science-datasets-main/Task 4/twitter_training.csv', header=None)
df.columns = ['ID', 'Entity', 'Sentiment', 'Text']
df.head()
```

## 2. Data Cleaning

```python
print('Original Shape:', df.shape)

# Drop duplicates
df = df.drop_duplicates()

# Drop missing values
df = df.dropna()

print('Shape after cleaning:', df.shape)
```

## 3. Exploratory Data Analysis & Visualization
### 3.1 Overall Sentiment Distribution

```python
plt.figure(figsize=(8,6))
sns.countplot(data=df, x='Sentiment', palette='viridis', order=df['Sentiment'].value_counts().index)
plt.title('Overall Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

### 3.2 Top Entities by Tweet Volume

```python
top_entities = df['Entity'].value_counts().head(10).index

plt.figure(figsize=(10,6))
sns.countplot(data=df[df['Entity'].isin(top_entities)], y='Entity', order=top_entities, palette='Set2')
plt.title('Top 10 Entities by Tweet Volume')
plt.xlabel('Count')
plt.ylabel('Entity')
plt.show()
```

### 3.3 Sentiment Distribution for Top Entities

```python
plt.figure(figsize=(14,8))
sns.countplot(data=df[df['Entity'].isin(top_entities)], y='Entity', hue='Sentiment', order=top_entities, palette='muted')
plt.title('Sentiment Distribution for Top 10 Entities')
plt.xlabel('Count')
plt.ylabel('Entity')
plt.legend(title='Sentiment', loc='lower right')
plt.show()
```

### 3.4 Word Clouds for Sentiments

```python
def generate_wordcloud(sentiment, title):
    text = ' '.join(str(tweet) for tweet in df[df['Sentiment'] == sentiment]['Text'])
    wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='inferno').generate(text)
    plt.figure(figsize=(10,5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(title)
    plt.show()

generate_wordcloud('Positive', 'Word Cloud - Positive Tweets')
generate_wordcloud('Negative', 'Word Cloud - Negative Tweets')
```

## 4. Insights

- The overall sentiment across all tweets shows a relatively balanced mix, with slightly more positive and negative tweets than neutral ones.
- Brands and games like `MaddenNFL`, `LeagueOfLegends`, and `CallOfDuty` are among the most discussed entities.
- The sentiment breakdown per entity reveals interesting brand-specific trends (e.g., some brands skew heavily negative while others skew positive).
- The word clouds highlight the most common terms associated with positive vs. negative feelings, which can help in understanding specific pain points or features users love.
