Load Dataset

```python
df = pd.read_csv("US_Accidents_March23.csv")
```

Basic Exploration

```python
print(df.head())

print(df.shape)

print(df.info())

print(df.isnull().sum())
```

Data Cleaning

```python
df = df.dropna(subset=['Weather_Condition'])

df['Start_Time'] = pd.to_datetime(df['Start_Time'])
```

Time-Based Analysis

Extract Hour and Day

```python
df['Hour'] = df['Start_Time'].dt.hour
df['Day'] = df['Start_Time'].dt.day_name()
```

Accidents by Hour

```python
plt.figure(figsize=(12,6))

sns.countplot(x=df['Hour'])

plt.title("Accidents by Hour")
plt.show()
```

Accidents by Day

```python
plt.figure(figsize=(12,6))

sns.countplot(x=df['Day'])

plt.title("Accidents by Day")
plt.xticks(rotation=45)
plt.show()
```

Weather Analysis

Top Weather Conditions

```python
weather_counts = df['Weather_Condition'].value_counts().head(10)

plt.figure(figsize=(12,6))

sns.barplot(x=weather_counts.values,
            y=weather_counts.index)

plt.title("Top Weather Conditions")
plt.show()
```

Severity Analysis

```python
plt.figure(figsize=(8,5))

sns.countplot(x=df['Severity'])

plt.title("Accident Severity Distribution")
plt.show()
```

Day vs Night Accidents

```python
sns.countplot(x=df['Sunrise_Sunset'])

plt.title("Day vs Night Accidents")
plt.show()
```

State-wise Accident Analysis

```python
top_states = df['State'].value_counts().head(10)

plt.figure(figsize=(12,6))

sns.barplot(x=top_states.index,
            y=top_states.values)

plt.title("Top 10 States by Accidents")
plt.show()
```
